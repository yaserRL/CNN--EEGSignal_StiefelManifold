```python
def rename_duplicate_columns(df):
    # Convert column names to strings to avoid type errors
    cols = pd.Series(df.columns.astype(str))
    
    # Iterate over duplicated columns and rename them
    for dup in cols[cols.duplicated()].unique():
        # Find duplicated columns and add a suffix to them
        cols[cols[cols == dup].index.values.tolist()] = [dup + f'_{i}' if i != 0 else dup for i in range(sum(cols == dup))]
    
    df.columns = cols
    return df

# Apply the function to each DataFrame
Gamma_Sub1_Vowel_hello_IA = rename_duplicate_columns(Gamma_Sub1_Vowel_hello_IA)
Gamma_Sub1_Vowel_helpme_IA = rename_duplicate_columns(Gamma_Sub1_Vowel_helpme_IA)
Gamma_Sub1_Vowel_stop_IA = rename_duplicate_columns(Gamma_Sub1_Vowel_stop_IA)
Gamma_Sub1_Vowel_thankyou_IA = rename_duplicate_columns(Gamma_Sub1_Vowel_thankyou_IA)
Gamma_Sub1_Vowel_yes_IA = rename_duplicate_columns(Gamma_Sub1_Vowel_yes_IA)
```
